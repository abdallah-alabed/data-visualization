# Data Visualization (Lab 14)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data1=sns.load_dataset('tips')
data2=sns.load_dataset('mpg')
```

## Dataset 1 (Tips)

#### this dataset has the data of bills and how much tips were given with other factors like (sex, smoking habit, time of the order and the table size)

**1. The relation between the total bill and tips from thursday till sunday**

```python
sns.set_theme(style="darkgrid")
plot1=sns.relplot(x="total_bill", y="tip", hue="day", data=data1)
plot1.set_axis_labels('total bill in dollars $','tips in dollars $')
```

##### We can see a variety in the tips when plotted against the total bill but we can conclude:
1. highest bills and highest tips are all on saturday
2. most of the tips are between 2 and 4 dollars

**2. The relation between the Gender and tips in lunch and dinner times**

```python
sns.relplot(x="sex", y="tip", hue="time", data=data1)
```

##### We can see a variety in the tips when plotted against the gender but we can conclude:
1. highest 3 tips are given by males and all of them are in the dinner time
2. males tips more on dinner time than lunch time
3. most tips are between 1 and 4 dollars

**3. The relation between the table size and tips**

```python
sns.relplot(x="size", y="tip",hue='day', data=data1)
```

##### We can see a variety in the tips when plotted against the size but we can conclude:
1. tables with 2,3 and 4 seats tips the most
2. size one table tip low
3. most tips are between 2 and 4 dollars

## Dataset 2 (mpg)

#### this dataset has the data of cars fuel consumption (miles per gallon) relative to the engine cylinders with other factors like (displacement, horsepower, weight and many more)

**1. The relation between the mpg for cars origin**

```python
sns.relplot(x="origin", y="mpg", data=data2)
```

##### We can conclude from the plot:
1. usa cars are the worst in mpg ratings
2. most economical car on the plot is japanese car

**2. The relation between the mpg and engine displacement**

```python
sns.relplot(x="displacement", y="mpg", data=data2)
```

##### We can conclude from the plot:
1. more engine displacement means less mpg rate
2. engines around 100 of displacement are the most economic

**3. The relation between the mpg and engine cylinders count**

```python
sns.relplot(x="cylinders", y="mpg",hue='displacement', data=data2)
```

##### We can conclude from the plot:
1. 4 cylinder engines are the most economical
2. 5 cylinder engines have different mpg values which is unusal
3. the larger the number of cylinders the lower mpg they have

**4. The relation between engine displacement and horsepower**

```python
sns.relplot(x="displacement", y="horsepower",hue='cylinders', data=data2)
```

##### We can conclude from the plot:
1. the more displacement the engine have the more horsepowerpower it generates
2. most engines with 100 displacement have horsepower between 50 and 100

**5. The relation between the acceleraation and horsepower**

```python
sns.relplot(x="horsepower",y="acceleration",hue='cylinders', data=data2)
```

##### We can conclude from the plot:
1. more horsepower means ower acceleration times (better)
2. only 8 cylinder vehicles were able to break the 10 second acceleration barrier

**6. Most Economical Vehicles**

```python
sorted=data2[['mpg','name']]
sorted=sorted.sort_values(by='mpg',ascending=False)
sorted.head(10)
```
